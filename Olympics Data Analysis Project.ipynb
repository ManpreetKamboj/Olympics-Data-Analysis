```python
#medals won by usa
def usamedals():
    g=data.loc[(data["Country"]=="USA")].groupby('Medal')
    m=g.size()
    print(g)
    root= tk.Tk() 

    figure1 = plt.Figure(figsize=(16,8), dpi=100)
    ax1 = figure1.add_subplot(111)
    bar1 = FigureCanvasTkAgg(figure1, root)
    bar1.get_tk_widget().pack(side=tk.LEFT, fill=tk.BOTH)
    m.plot(kind='bar', ax=ax1)
    ax1.set_title('Medals won by USA')
    root.mainloop()
```

```python
#plotting of medals won yearwise
def medalywise():
    data1=data.loc[(data["Country"]=='USA')].groupby(data['Year']).size()
    root= tk.Tk() 

    figure1 = plt.Figure(figsize=(16,8), dpi=100)
    ax1 = figure1.add_subplot(111)
    bar1 = FigureCanvasTkAgg(figure1, root)
    bar1.get_tk_widget().pack(side=tk.LEFT, fill=tk.BOTH)
    data1.plot(kind='bar', ax=ax1)
    ax1.set_title('Total Medals Yearwise')
    root.mainloop()
```
